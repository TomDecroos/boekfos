```python
from boekfos.core import Plan,Rect,Box
from boekfos.space import Room, Floor, Garden
from boekfos.outside import Carpark, Terrace, Roof
from boekfos.furniture import Closet, Furniture, Table
from boekfos.orientation import Side
```

```python
p = Plan()
f = Floor(Box(0,0,8,15))
p.add(f)
garden = Garden(f.outer.gobottom().edit(-4,-6,4,42-6)) 
p.add(garden)
garden.bottomwall.hide()
garden.rightwall.hole(0,1.25)
garden2 = Garden(f.outer.gobottom().edit(4.5,-4.5,4,-1),grass=False)
p.add(garden2)
garden2.topwall.hide()
garden2.rightwall.hide()
carpark = Carpark(Box(-4,-6,4.5,-1))
p.add(carpark)
carpark.setlevel(11)
carpark2 = Carpark(Box(4.5,-6,12,-4.5))
p.add(carpark2)
garden.ground.setlevel(12)
p.add(Terrace(f.outer.edit(-1,-1,1,3)))

f.topwall.window(3,-0.25)
f.rightwall.window(8,-3)

# KEUKEN
keukenkast1 = f.leftwall.closet(Side.RIGHT,0.5,-5.2,0)
p.add(keukenkast1)

keukenkast2 = f.topwall.closet(Side.BOTTOM,0.5,0,2.96)
p.add(keukenkast2)

keukenkast3 = f.topwall.closet(Side.BOTTOM,1,1.96,2.96)
p.add(keukenkast3)

#kookeiland = Table(keukenkast1.gotop().goright().edit(1.2,-4,1.2+1,-0.7))
kookeiland = Table(Box(keukenkast3.left,keukenkast3.bottom-3,keukenkast3.right,keukenkast3.bottom))
vuur = Furniture(kookeiland.boundingbox().gotop().edit(0.1,-0.2-0.7,-0.3,-0.2))
afwasbak1 = Furniture(kookeiland.boundingbox().gobottom().edit(0.1,0.8,-0.3,0.8+0.4))
afwasbak2 = Furniture(afwasbak1.edit(0,0.6,0,0.6))
p.add(kookeiland)
p.add(vuur)
p.add(afwasbak1)
p.add(afwasbak2)
p.add(kookeiland.chair(Side.LEFT,0,7))
p.add(kookeiland.chair(Side.RIGHT,0,7))



tafel = Table(f.inner.gotop().goright().edit(0,-1,0,-1).edit(-2.2,-1,0,0))
p.add(tafel)
for i in range(4):
    p.add(tafel.chair(Side.TOP,i,4))
    p.add(tafel.chair(Side.BOTTOM,i,4))

# TOILET + BADKAMER
toilet = Room(f.inner.gobottom().goleft().edit(-0.05,0,1.75,1))
toilet.leftwall.hide()
toilet.bottomwall.hide()
p.add(toilet)
p.add(toilet.leftwall.toilet(Side.RIGHT))
p.add(toilet.bottomwall.sink(Side.TOP,-0.5,-0.05))

badkamer = Room(toilet.center.gotop().edit(0,0,0,2),text="Badkamer")
badkamer.leftwall.hide()
p.add(badkamer)
p.add(badkamer.leftwall.sink(Side.RIGHT,0.55,0.1))
p.add(badkamer.leftwall.table(Side.RIGHT,0.5,0,1))
p.add(badkamer.topwall.shower(Side.RIGHT,Side.BOTTOM,0))


# HAL
hal = Room(Box(badkamer.center.right,f.inner.bottom,f.inner.right-2.5,f.inner.bottom + 2.5),text="Hal")
hal.leftwall.hide()
hal.topwall.hole(0,0.95)
hal.bottomwall.hide()
f.add(hal)

p.add(hal.rightwall.closet(Side.LEFT,0.5,0,-0.05))

## BERGRUIMTE
berging = Room(Box(f.inner.left,hal.center.top,f.inner.left+2.75,keukenkast1.bottom),text="Berging")
berging.leftwall.hide()
berging.bottomwall.hide()
p.add(berging)
p.add(berging.rightwall.closet(Side.LEFT,0.5,3.5,0))
p.add(f.leftwall.shower(Side.BOTTOM,Side.RIGHT,-5-0.5))
p.add(f.leftwall.shower(Side.BOTTOM,Side.RIGHT,-5-1.5))
p.add(f.leftwall.shower(Side.BOTTOM,Side.RIGHT,-5-2.5))
#keukenkast2 = berging.topwall.closet(Side.TOP,0.5,0,2)
#p.add(keukenkast2)



bureau = Room(Box(hal.center.right,f.inner.bottom,f.inner.right,hal.center.top+1),text='Bureau')
p.add(bureau)

# SOFA, TV, BUREAU, LIVINGKAST
zetelbottom = bureau.outer.top + 0.1
zetelkorttop = zetelbottom + 0.9
zetellangtop = zetelbottom + 3.50
zetelright = f.inner.right - 0.1
zetellangleft = zetelright - 0.9
zetelkortleft = zetelright - 2.5
sofa1 = Furniture(Box(zetelkortleft,zetelbottom,zetellangleft,zetelkorttop))
sofa2 = Furniture(Box(zetellangleft,zetelkorttop,zetelright,zetellangtop))
sofatafel = Furniture(Box(zetelkortleft+0.5,zetelkorttop+0.5,zetellangleft-0.5,zetellangtop-0.5))
p.add(sofa1)
p.add(sofa2)
p.add(sofatafel)

p.add(bureau.leftwall.bureau(Side.RIGHT,0,-1.5))
p.add(bureau.rightwall.closet(Side.LEFT,0.5,0,0))


livingkast = berging.rightwall.closet(Side.RIGHT,0.4,2,0)
p.add(livingkast)




stairs = berging.bottomwall.stairs(Side.TOP,-0.95,-0.05)
p.add(stairs)

berging.topwall.door(Side.RIGHT,Side.BOTTOM,1.1)
toilet.rightwall.door(Side.TOP,Side.RIGHT,0.1)
badkamer.rightwall.door(Side.TOP,Side.LEFT,0.1)
hal.topwall.door(Side.RIGHT,Side.TOP,-1.5)
f.leftwall.door(Side.BOTTOM,Side.RIGHT,6,1)
f.bottomwall.door(Side.LEFT,Side.TOP,4.3,1)
bureau.leftwall.door(Side.BOTTOM,Side.RIGHT,-0.1)


print(berging.center.surface())
p.draw(figsize=0.8)
```
