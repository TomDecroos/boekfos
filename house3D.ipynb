```python
%matplotlib notebook
"""
============================
Draw flat objects in 3D plot
============================

Demonstrate using pathpatch_2d_to_3d to 'draw' shapes and text on a 3D plot.
"""

import matplotlib.pyplot as plt
from matplotlib.patches import Circle, PathPatch,Rectangle
# register Axes3D class with matplotlib by importing Axes3D
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
from matplotlib.text import TextPath
from matplotlib.transforms import Affine2D


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

def cube(p1,p2):
    x1,y1,z1 = p1
    x2,y2,z2 = p2
    w = x2-x1
    b = y2-y1
    h = z2 - z1
    
    ps = [Rectangle(x1,y1,6)]


def zrectangle(x1,y1,x2,y2,z,color=None):
    p = Rectangle((x1, y1), (x2-x1),(y2-y1),color=color)
    ax.add_patch(p)
    art3d.pathpatch_2d_to_3d(p, z=z, zdir="z")

def xrectangle(y1,z1,y2,z2,x,color=None):
    p = Rectangle((y1, z1), (y2-y1),(z2-z1),color=color)
    ax.add_patch(p)
    art3d.pathpatch_2d_to_3d(p, z=x, zdir="x")
def yrectangle(x1,z1,x2,z2,y,color=None):
    p = Rectangle((x1, z1), (x2-x1),(z2-z1),color=color)
    ax.add_patch(p)
    art3d.pathpatch_2d_to_3d(p, z=y, zdir="y")

def bar(p1,p2,color=None):
    x1,y1,z1 = p1
    x2,y2,z2 = p2
    zrectangle(x1,y1,x2,y2,z1,color)
    zrectangle(x1,y1,x2,y2,z2,color)
    xrectangle(y1,z1,y2,z2,x1,color)
    xrectangle(y1,z1,y2,z2,x2,color)
    yrectangle(x1,z1,x2,z2,y1,color)
    yrectangle(x1,z1,x2,z2,y2,color)

def xframe(y1,z1,y2,z2,x,color="black"):
    for y in [y1,y2]:
        ax.plot([y,y],[z1,z2],zs=x,zdir="x",color="black",lw=3)
    for z in [z1,z2]:
        ax.plot([y1,y2],[z,z],zs=x,zdir="x",color="black",lw=3)

def yframe(y1,z1,y2,z2,x,color="black"):
    for y in [y1,y2]:
        ax.plot([y,y],[z1,z2],zs=x,zdir="y",color="black",lw=3)
    for z in [z1,z2]:
        ax.plot([y1,y2],[z,z],zs=x,zdir="y",color="black",lw=3)

def zframe(y1,z1,y2,z2,x,color="black"):
    for y in [y1,y2]:
        ax.plot([y,y],[z1,z2],zs=x,zdir="z",color="black",lw=3)
    for z in [z1,z2]:
        ax.plot([y1,y2],[z,z],zs=x,zdir="z",color="black",lw=3)

def barframe(p1,p2,color="black"):
    x1,y1,z1 = p1
    x2,y2,z2 = p2
    zframe(x1,y1,x2,y2,z1,color)
    zframe(x1,y1,x2,y2,z2,color)
    xframe(y1,z1,y2,z2,x1,color)
    xframe(y1,z1,y2,z2,x2,color)
    yframe(x1,z1,x2,z2,y1,color)
    yframe(x1,z1,x2,z2,y2,color)

barframe((10,0,0),(18,18,4),"grey")
barframe((10,0,4),(18,12,8),"grey")

def ground(x1,y1,x2,y2,z=0,thickness=0.4):
    pass

ax.set_xlim3d(-1,18+1)
ax.set_ylim3d(-1,18+1)
ax.set_zlim3d(-1,18+1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```
